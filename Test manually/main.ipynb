```python
import numpy as np

# Example dataset
x = np.array([1.1, 1.3, 1.5, 2.0, 2.2, 2.9, 3.0, 3.2, 3.2, 3.7])  # Years of Experience
y = np.array([39343, 46205, 37731, 43525, 39891, 56642, 60150, 54445, 64445, 57189])  # Salary

# Step 1: Calculate the mean of x and y
mean_x = np.mean(x)
mean_y = np.mean(y)

# Step 2: Calculate the slope (m)
numerator = np.sum((x - mean_x) * (y - mean_y))  # Sum of (x_i - x̄)(y_i - ȳ)
denominator = np.sum((x - mean_x) ** 2)         # Sum of (x_i - x̄)^2
m = numerator / denominator

# Step 3: Calculate the intercept (b)
b = mean_y - m * mean_x

# Step 4: Make predictions
def predict(x):
    return m * x + b

# Example prediction
new_x = 5.0  # 5 years of experience
predicted_y = predict(new_x)
print(f'Predicted Salary for {new_x} years of experience: {predicted_y:.2f}')

# Output the slope and intercept
print(f'Slope (m): {m:.2f}')
print(f'Intercept (b): {b:.2f}')
```

```python
import matplotlib.pyplot as plt

# Plot the data points
plt.scatter(x, y, color='blue', label='Data Points')

# Plot the regression line
regression_line = m * x + b
plt.plot(x, regression_line, color='red', label='Regression Line')

# Add labels and legend
plt.title('Salary vs Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
